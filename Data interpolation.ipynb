## Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:

#1. Linear Interpolation

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
#interpolate the data using linear interplotion
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
#interpolate the data using linear interplotion
x_new = np.linspace(1,5,10) ##create new x values
y_interp = np.interp(x_new, x,y)##interplate y values
print(y_interp)
```

```python
plt.scatter(x_new, y_interp)
```

Cubic Interpolation with scipy

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
##cretae a  cubic interpolation function
f = interp1d(x,y,kind = 'cubic')
```

```python
#interpolate the data
x_new = np.linspace(1,5,10)
y_interp = f(x_new)
print(y_interp)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new, y_interp)
```

Polynomial interpolation

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,9,16,25])
```

```python
# interpolate the data using polynomial interpolation
p = np.polyfit(x,y,4) # fit a 2nd degreee polynomial to the data
```

```python
#interpolate the data using polynomial interplotion
x_new = np.linspace(1,5,10) ##create new x values
y_interp = np.polyval(p, x_new)##interplate y values
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
